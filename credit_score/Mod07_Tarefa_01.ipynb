#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
quali = df.select_dtypes(include=['object']).columns.tolist()
quali
````

````python
df_dummies = pd.get_dummies(df, columns=quali, drop_first=True)
df_dummies.head()
````

````python
df_dummies.info()
````

````python
boleanos = df_dummies.select_dtypes(include=['bool']).columns

for a in boleanos:
    df_dummies[a] = df_dummies[a].map({True: 1, False: 0})
df_dummies.info()
````

````python
df_dummies.head()
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
cross_tab_email = pd.crosstab(df_dummies['possui_email'], df_dummies['mau'], normalize='index')
cross_tab_veiculo = pd.crosstab(df_dummies['posse_de_veiculo_Y'], df_dummies['mau'], normalize='index')
````

````python
cross_tab_email
````

````python
cross_tab_veiculo
````

Ambas as variáveis têm uma distribuição semelhante em relação à probabilidade de mau. No entanto, a variável possui_email apresenta uma ligeira diferença maior entre as categorias (0 e 1) em comparação com posse_de_veiculo. Portanto, possui_email pode ser ligeiramente mais poderosa para prever a probabilidade de mau = 1.

* Mas se for aproundar um pouco mais da pra tentar olhar pelo p valor dos qui-quadrado

````python
import scipy.stats as stats

# Perform chi-square test for independence for 'possui_email' and 'mau'
chi2_email, p_email, dof_email, ex_email = stats.chi2_contingency(pd.crosstab(df_dummies['possui_email'], df_dummies['mau']))

# Perform chi-square test for independence for 'posse_de_veiculo' and 'mau'
chi2_veiculo, p_veiculo, dof_veiculo, ex_veiculo = stats.chi2_contingency(pd.crosstab(df_dummies['posse_de_veiculo_Y'], df_dummies['mau']))

p_email, p_veiculo
````

* **P-valores dos testes qui-quadrado:**

possui_email:
p=0.453
p=0.453
posse_de_veiculo:
p=0.395
p=0.395

Os p-valores para ambas as variáveis são maiores que 0.05, indicando que não há evidência estatística suficiente para afirmar que há uma associação significativa entre essas variáveis e a probabilidade de mau = 1.

Com base nos p-valores, nenhuma das variáveis parece ser uma preditora poderosa da probabilidade de mau = 1.
